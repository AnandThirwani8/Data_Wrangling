# Buckets

```python
d['sales'].describe()
```

```python
# Create a function that will take row as an input

def f(row):
    if row['sales'] >= 0 and row['sales'] < 10:
        val = '0-10'
    elif row['sales'] >= 10 and row['sales'] < 20:
        val = '20-30'
    else:
        val = '30+'
    return val
```

```python
# it is important to mention axis = 1 here

d['buckets'] = d.apply(f, axis=1)
```

```python
d.head()
```

# Grouping

Always to reset_index() after using group by

```python
# Let's see the distribution of sales in each bucket

d.groupby('buckets')['sales'].describe().reset_index()
```

```python
# filter to get a smaller dataframe

condition = (d['item'].isin([55732362,56285036])) & (d['store_nbr'].isin([1014,1025]))
d = d[condition]
```

```python
# Group by multiple columns
# reset_index() also allows us to name the newly created column
d_item = d[(d['item'] == 55732362)]

d.groupby(['item','store_nbr'])['sales'].sum().reset_index(name = 'sum_of_sales')
```

```python
# Exactly similar thing can be dome using 'agg' method

d.groupby(['item','store_nbr'])['sales'].agg('sum').reset_index(name = 'sum_of_sales')
```

```python
# Advantage of agg function is that multiple aggregations can be done after group by

# to get count(*) of a group, just write 'count'

d.groupby(['item','store_nbr'])['sales'].agg(['sum', np.mean, 'count']).reset_index()
```

```python
# Agg function is very versatile

d.groupby(['item','store_nbr']).agg({
                                     'sales':['mean', 'sum'], 
                                     'units':['mean', 'sum'],
                                    }).reset_index()
```

# Dplyr Mutate

```python
# create a small df of 3 items

# intialise data of lists. 
data = { 'Nationality':['India','India','US','US'], 
         'Name':['Mac', 'Rahul','Nick', 'Jack'], 
         'Age':[20, 16, 21, 19] } 
   
# Create DataFrame 
df = pd.DataFrame(data) 
  
# Print the output. 
df 
```

```python
# Group by and create mean and sum

df.groupby('Nationality')['Age'].agg(['mean','sum']).reset_index()
```

```python
# Show mean and sum of the group with the entire datafarme

df['mean'] = df.groupby('Nationality')['Age'].transform('mean')

df['sum'] = df.groupby('Nationality')['Age'].transform('sum')

df
```

# Self Join - Cartesian Product

```python
# create a small df of 3 items

# intialise data of lists. 
data = { 'Name':['Mac', 'Nick', 'Jack'], 'Age':[20, 21, 19] } 
   
# Create DataFrame 
df = pd.DataFrame(data) 
  
# Print the output. 
df 
```

```python
# We want to create a dataframe with 9 rows. Each row with a unique pair of items
```

```python
# create a dummy column to join on
df['key'] = 1
```

```python
df
```

```python
# do a simple join
# then drop the key column (optional)

pd.merge(df,df, on='key').drop('key', axis='columns')
```
